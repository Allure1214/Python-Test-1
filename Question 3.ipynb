```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a DataFrame
coffee_df = pd.read_csv('coffee.csv')

# Group by 'product_category' and 'product_type', then sum the 'transaction_qty'
grouped_df = coffee_df.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

# Rename the column to 'total_order' to match the table in the image
grouped_df.rename(columns={'transaction_qty': 'total_order'}, inplace=True)

# Display the DataFrame
print(grouped_df)

# Set the plot size
plt.figure(figsize=(12, 8))

# Create a bar plot to visualize the total orders
sns.barplot(x='total_order', y='product_type', hue='product_category', data=grouped_df, dodge=False)

# Add labels and title
plt.xlabel('Total Orders')
plt.ylabel('Product Type')
plt.title('Total Orders by Product Type and Category')

# Show plot
plt.show()
```
